# Graphs

Data structure designed to show relationships between objects. Objects we defined as nodes in past, with multiple properties. AKA Network. Similar to tree, they have nodes or vertices and the relationships are edges. A tree is a specific type of graph. Graphs do not have a root node, because they have cycles. Edges can store data too? Like a number, representing # of interactions.

Edges can have a directions. Directed graph means graph where the edges have a direction. It is essentially a pointer. DC -> Philly. They can also go both ways DC <-> Philly. Cycles can lead to infiite loops! Make sure graph is acyclic for algorithms. One solution: DAG - Directed Acyclic Graph. 

Connectivity has a specific meaning in graph theory. Disconnected graph is a vertex with no edges to the other vertices. A connected graph has no disconnected vertices. Measures the min number of elements that need to be removed for a graph to become disconnected. Can help determine which graph is stronger. 

- **Disconnected**: no connection to graph
- **Weakly Connected**: if directed edges are replaced with undirected edges, the graph becomes connected. For example, if a vertex has a directed edge pointing to rest of data, but no other vertex points to it. 
- **Connected**: Only used to refer to undirected graphs 
- **Strongly Connected**: Must have path from every node to every other node

## Graph Representations

Can make vertex and edges as objects. Can be tough to traverse graph if need to search through objects. 

**Edge List**: a list of edges. Edges are represented by a list of 2 elements, aka their vertices.[[0,1], [1,2], [1,3]].

**Adjacency List**: Vertices have an id # that corresponds to index in the array. In the array, each space stores a list of nodes that the index is adjacent to. 2-dimensional list, because it is a list that contains another list. 

**Adjacency Matrix**: 2 D array, but the lists inside are all the same length. Similar to adjacency list. Indices still represent adjacent nodes, however it is binary. 1 for connection/edge or 0 for no connectivity to that particular node.

```python
# Graphs Practice

class Node(object):
    def __init__(self, value):
        self.value = value
        self.edges = []

class Edge(object):
    def __init__(self, value, node_from, node_to):
        self.value = value
        self.node_from = node_from
        self.node_to = node_to

class Graph(object):
    def __init__(self, nodes=[], edges=[]):
        self.nodes = nodes
        self.edges = edges

    def insert_node(self, new_node_val):
        new_node = Node(new_node_val)
        self.nodes.append(new_node)
        
    def insert_edge(self, new_edge_val, node_from_val, node_to_val):
        from_found = None
        to_found = None
        for node in self.nodes:
            if node_from_val == node.value:
                from_found = node
            if node_to_val == node.value:
                to_found = node
        if from_found == None:
            from_found = Node(node_from_val)
            self.nodes.append(from_found)
        if to_found == None:
            to_found = Node(node_to_val)
            self.nodes.append(to_found)
        new_edge = Edge(new_edge_val, from_found, to_found)
        from_found.edges.append(new_edge)
        to_found.edges.append(new_edge)
        self.edges.append(new_edge)

    def get_edge_list(self):
        """Don't return a list of edge objects!
        Return a list of triples that looks like this:
        (Edge Value, From Node Value, To Node Value)"""
        list_of_edges = []
        for edge in self.edges:
            list_of_edges.append([edge.value, edge.node_from.value, edge.node_to.value])
        return list_of_edges

    def get_adjacency_list(self):
        """Don't return any Node or Edge objects!
        You'll return a list of lists.
        The indecies of the outer list represent
        "from" nodes.
        Each section in the list will store a list
        of tuples that looks like this:
        (To Node, Edge Value)"""
        # Return edges that are pointing to another node
        # Get total number of nodes
        # adj_list = {node: [] for node in range(5)}
        # for edge in self.edges:
        #     # adj_list.insert(1, 2)
        #     adj_list[edge.node_from.value].append({edge.node_to.value, edge.value})
        # return adj_list

        count = 1
        for i in self.nodes:
            count += 1
        adj_list = [[] for i in range(count)]
        for edge in self.edges:
            adj_list[edge.node_from.value].append((edge.node_to.value, edge.value))
        for i in range(count):
            if adj_list[i] == []:
                adj_list[i] = None
        return adj_list

    
    def get_adjacency_matrix(self):
        """Return a matrix, or 2D list.
        Row numbers represent from nodes,
        column numbers represent to nodes.
        Store the edge values in each spot,
        and a 0 if no edge exists."""
        # Built matrix
        num = 1
        for i in self.nodes:
            num += 1
        matrix = [0] * num
        for x in range(num):
            matrix[x] = [0] * num
        for edge in self.edges:
            matrix[edge.node_from.value][edge.node_to.value] = edge.value
        
        return matrix

graph = Graph()
graph.insert_edge(100, 1, 2)
graph.insert_edge(101, 1, 3)
graph.insert_edge(102, 1, 4)
graph.insert_edge(103, 3, 4)
# Should be [(100, 1, 2), (101, 1, 3), (102, 1, 4), (103, 3, 4)]
print(graph.get_edge_list())
# Should be [None, [(2, 100), (3, 101), (4, 102)], None, [(4, 103)], None]
print(graph.get_adjacency_list())
# Should be [[0, 0, 0, 0, 0], [0, 0, 100, 101, 102], [0, 0, 0, 0, 0], [0, 0, 0, 0, 103], [0, 0, 0, 0, 0]]
print(graph.get_adjacency_matrix())
```

# Graph Traversal

Graphs are great for modeling connections between elements, and are easy to traverse. Tree is a specific type of graph.

**Depth First Search (DFS)** - No root in a graph, begin at any node. Same principles as on trees. Mark seen nodes under a stack. As you see more nodes, add to stack. If you see a node you've already seen, go back to node and try another edge. Exclusively vising all edges and vertices at least once. Time to traverse is still O(E + V), O(2n), but can vary based on implementation. 

**Bredth First Search (BFS)** - Similar, still visiting every edge and marking off every node. Searching every edge of one node, before moving on. Add seen nodes connected to iniital node to Queue. Dequeue node from queue and use that as next node. Efficiency is still O(E + V).

# Notable Paths

- **Eulerian Paths**: Path that traverses each edge of a graph once, and may end up at a different node. For Eulerien Cycles you must traverse every node only once, and end up at original node. Graphs can only have Eulerian cycles if all vertices have an even degree, or an even number of edges connected to the node. O(n), go through each edge once.
- **Hamiltonian Paths**: Another type of path that must go through every vertex once. It will start and end with the same vertex.
