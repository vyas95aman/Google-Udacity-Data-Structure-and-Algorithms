Linked lists are horizonal, while trees are drawn vertically. The individual cells in a tree are called nodes. Trees do not have any unconnected nodes and do not have cycles!
Parents are the nodes that have children. Children can only have one parent. Nodes that do not have children are called leaves. Height of a tree is the height of the root node starting at 0 (LEAF TO ROOT). Depth is the number of nodes to the root starting at 0 (ROOT TO LEAF). LEVELS are how many connections it takes to reach the root, and root is always 1. 

### Traversal:
- **DFS** - Depth First Search - explore all children first
- **BFS** - Bredth First Search - explore current level first then move to child level - Level-Order Traversal 

### DFS:
- **Pre Order Traversals** - check off node as soon as you see it before traversing down. Move down left most nodes until you hit a leaf. Then go to parent and go to right child. 
- **In Order Traversals** - Move through the same way, but check off differently. Once we see the left child, check off then come back to parent and check off. 
- **Post Order Traversal** - Can't check off parent until see all of its children aka leaves. Left leaf check, parent,  right leaf check, then parent check

Searching O(n), linear search.
Deleting O(n), can replace deleted node with a leaf.

Inserting nodes - each row can double the ammount of nodes in current level.

Binary Search Trees are binary trees, but they are sorted. Left of node is smaller, right of node is larger. Can search, insert and delete quickly. Runtime on a BST is the same as the Height of the tree, aka O(log(n)). 

Nice when they are full, but can end up unbalanced. 

```python
# Binary Search Tree Practice - Insert and Search

class Node(object):
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST(object):
    def __init__(self, root):
        self.root = Node(root)

    def insert(self, node, new_val):
        if node is None:
            return Node(new_val)
        else:
            if new_val < node.value:
                node.left = self.insert(node.left, new_val)
            if new_val > node.value:
                node.right = self.insert(node.right, new_val)
            return node
    

    def search(self, node, find_val):
        if node is None:
            return False
        if node.value == find_val:
            return True
        if find_val < node.value:
            return self.search(node.left, find_val)
        else:
            return self.search(node.right, find_val)
            
# Set up tree
tree = BST(4)

# Insert elements
tree.insert(tree.root, 2)
tree.insert(tree.root, 1)
tree.insert(tree.root, 3)
tree.insert(tree.root, 5)

# Check search
# Should be True
print(tree.search(tree.root, 4))
# Should be False
print(tree.search(tree.root, 6))

# My checks
print(tree.root.value)
print(tree.root.left.value)
```

**Heaps:**

Heap is a specific kind of tree. Elements are arranged in increasing or decreasing order, root is either max or min -> max heap and min heap. Max heap = parent always has greater value than its child. *Do not need to be binary trees. Heap trees must be complete trees (all levels full besides last). Max value, aka peak, happens in constant time. Searching is a linear time operation, O(n), because can't decide which child to go to. 

**Heapify:**

To add a number to a Heap tree, we add number to open node, then heapify. Heapify reorders tree based on heap property. Keep comparing new element with parent and swap when bigger. Inserting/Deleting will be worst case O(log(n)).

**Heap Implementation:**

Represented as trees, but actually often stored as arrays. This only works with arrays sorted in descending order. 

**Self-Balancing Trees**

Use of an algorithm to minimize the number of levels used. Most common example is a Red-Black Tree. Nodes are assigned an additional color property to distinguish nodes. The Leaves also have null leaf nodes. Root is black, only if two nulls is the node Red. Insert red nodes. The number of black nodes must be the same on both sides.
Insert is O(log(n)) in worst and average cases.
